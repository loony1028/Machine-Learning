`Converting Labels to Numeric by Direct Mapping`

```python
spam_detection['Label'] = spam_detection['Label'].map({
    'ham': 0,
    'spam': 1
})
```

`Text Preprocessing Using NLTK`

```python
nltk.download('stopwords')
```

`Defining Data Cleaning Function`

```python
ps = PorterStemmer()

def preprocess_text(text):
    text = re.sub('[^a-zA-Z]', ' ', text)
    text = text.lower()
    words = text.split()
    words = [ps.stem(word) for word in words if word not in set(stopwords.words('english'))]
    return ' '.join(words)
```

```python
spam_detection['Clean_Message'] = spam_detection['Message'].apply(preprocess_text)
```
